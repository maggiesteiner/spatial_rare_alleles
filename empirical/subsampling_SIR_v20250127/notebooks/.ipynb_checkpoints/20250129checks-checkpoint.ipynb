```python
dfpca_lof = pd.read_csv('../results/daf/chr1_lof_uniformpca_nSIR10000_nSIRreps10.daf',sep='\t')
fixed_alleles_lof = dfpca_lof.groupby('POS').filter(lambda group: (group['Derived_Frequency'] == 1.0).all())
fixed_alleles_lof = fixed_alleles_lof['POS'].unique()
fixed_alleles_lof
```

```python
dfpca_mis = pd.read_csv('../results/daf/chr1_missense_uniformpca_nSIR10000_nSIRreps10.daf',sep='\t')
fixed_alleles_mis = dfpca_lof.groupby('POS').filter(lambda group: (group['Derived_Frequency'] == 1.0).all())
fixed_alleles_mis = fixed_alleles_mis['POS'].unique()
fixed_alleles_mis
```

```python
dfpca_syn = pd.read_csv('../results/daf/chr1_synonymous_uniformpca_nSIR10000_nSIRreps10.daf',sep='\t')
fixed_alleles_syn = dfpca_lof.groupby('POS').filter(lambda group: (group['Derived_Frequency'] == 1.0).all())
fixed_alleles_syn = fixed_alleles_syn['POS'].unique()
fixed_alleles_syn
```

```python
def process_files(pattern, fixed_alleles, label, color):
    files = glob.glob(pattern)
    dfs = {file.split('/')[-1]: pd.read_csv(file, delimiter='\t') for file in files}
    
    mean_freqs = {}
    for file, df in dfs.items():
        scenario = file.split('_')[2]
        df_filtered = df[~df['POS'].isin(fixed_alleles)]
        mean_freq = df_filtered.groupby('ITER')['Derived_Frequency'].mean()
        mean_freqs[scenario] = mean_freq
    
    mean_freqs_df = pd.DataFrame(mean_freqs)
    means = mean_freqs_df.mean()
    std_devs = mean_freqs_df.std()

    custom_order = [
        'centerE16N4geo50000', 'centerE16N4geo100000', 'centerE16N4geo150000',
        'centerE9N9geo50000', 'centerE9N9geo100000', 'centerE9N9geo150000',
        'centerE6N4geo50000', 'centerE6N4geo100000', 'centerE6N4geo150000',
        'uniformgeo'#,
        #'centerX19Y4pca0.0015', 'centerX19Y4pca0.0025', 'centerX19Y4pca0.005',
        #'uniformpca'
    ]
    
    means = means[custom_order]
    std_devs = std_devs[custom_order]
    print(means.iloc[-1])
    
    percentage_change = ((means.iloc[-1] - means.iloc[0]) / means.iloc[0]) * 100
    print(f"{label} Percentage change (uniform vs E16N4 narrow: {percentage_change:.4f}%")
    percentage_change = ((means.iloc[-1] - means.iloc[3]) / means.iloc[3]) * 100
    print(f"{label} Percentage change (uniform vs E9N9 narrow: {percentage_change:.4f}%")
    percentage_change = ((means.iloc[-1] - means.iloc[6]) / means.iloc[6]) * 100
    print(f"{label} Percentage change (uniform vs E6N4 narrow: {percentage_change:.4f}%")
    # percentage_change = ((means.iloc[-1] - means.iloc[-4]) / means.iloc[-4]) * 100
    # print(f"{label} Percentage change (pca0.0015 vs uniformpca): {percentage_change:.4f}%")
    # percentage_change = ((means.iloc[2] - means.iloc[0]) / means.iloc[0]) * 100
    # print(f"{label} Percentage change (centerE16N4geo50000 vs centerE16N4geo150000): {percentage_change:.4f}%")
    
    return means, std_devs, label, color

plt.figure(figsize=(10, 6))

# Process each category
lof_means, lof_stds, lof_label, lof_color = process_files('../results/daf/chr1_lof*geo*', fixed_alleles_lof, 'LOF', 'blue')
mis_means, mis_stds, mis_label, mis_color = process_files('../results/daf/chr1_mis*geo*', fixed_alleles_mis, 'MIS', 'red')
syn_means, syn_stds, syn_label, syn_color = process_files('../results/daf/chr1_syn*geo*', fixed_alleles_syn, 'SYN', 'green')

# Plot all data
plt.errorbar(lof_means.index, lof_means, yerr=lof_stds, fmt='o', capsize=5, color=lof_color, ecolor='black', label=lof_label)
plt.errorbar(mis_means.index, mis_means, yerr=mis_stds, fmt='o', capsize=5, color=mis_color, ecolor='black', label=mis_label)
plt.errorbar(syn_means.index, syn_means, yerr=syn_stds, fmt='o', capsize=5, color=syn_color, ecolor='black', label=syn_label)

# Final plot adjustments
plt.xticks(rotation=90)
plt.title("Avg. Derived allele frequencies")
plt.ylabel("Avg. Derived allele frequencies")
plt.legend()
plt.tight_layout()
plt.yscale('log')
plt.ylim(1e-4,1e-2)
plt.show()
# plt.savefig('derived_freq_v2.pdf')
```

```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import numpy as np

def plot_sfs_for_scenarios(pattern, fixed_alleles, label):
    files = glob.glob(pattern)
    dfs = {file.split('/')[-1]: pd.read_csv(file, delimiter='\t') for file in files}
    
    all_sfs = {}
    
    # First pass: gather all SFS data
    for file, df in dfs.items():
        scenario = file.split('_')[2]
        df_filtered = df[~df['POS'].isin(fixed_alleles)]
        
        # Calculate derived allele counts for each iteration
        df_filtered['Derived_Alleles_Counts'] = (df_filtered['Derived_Frequency'] * 2 * df_filtered['NOBS']).round().astype(int)
        
        # Calculate SFS for each scenario and iteration
        iter_sfs = df_filtered.groupby(['ITER', 'Derived_Alleles_Counts'])['Derived_Alleles_Counts'].count().unstack(fill_value=0)
        all_sfs[scenario] = iter_sfs

    # Handle uniformgeo separately
    dfunif = pd.read_csv('../results/daf/chr1_lof_uniformgeo_nSIR10000_nSIRreps10.daf', delimiter='\t')
    dfunif_filtered = dfunif[~dfunif['POS'].isin(fixed_alleles)]
    dfunif_filtered['Derived_Alleles_Counts'] = (dfunif_filtered['Derived_Frequency'] * 2 * dfunif_filtered['NOBS']).round().astype(int)
    iter_sfs_unif = dfunif_filtered.groupby(['ITER', 'Derived_Alleles_Counts'])['Derived_Alleles_Counts'].count().unstack(fill_value=0)

    # Define color mapping for different widths
    width_colors = {
        '50000': 'blue',
        '100000': 'red',
        '150000': 'green',
        'uniformgeo': 'purple'
    }

    # Plotting the SFS for each center in separate panels
    fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

    centers = ['centerE16N4', 'centerE9N9', 'centerE6N4']

    for i, center in enumerate(centers):
        ax = axes[i]
        # Plot for each width
        for width in ['50000', '100000', '150000']:
            file = f'{center}geo{width}'
            iter_sfs_counts = all_sfs[file]
            mean_sfs = iter_sfs_counts.mean(axis=0)
            std_sfs = iter_sfs_counts.std(axis=0)

            # Plot mean and stdev as error bars with loglog scale
            ax.loglog(mean_sfs.index, mean_sfs.values, marker='o', linestyle='-', 
                      label=f'{width} {center}', color=width_colors[width])
            ax.errorbar(mean_sfs.index, mean_sfs.values, yerr=std_sfs.values, fmt='o', color=width_colors[width], 
                        capsize=5, linestyle='None', alpha=0.7)
        
        # Handle uniformgeo separately, ensuring the correct label and color
        mean_sfs_unif = iter_sfs_unif.mean(axis=0)
        std_sfs_unif = iter_sfs_unif.std(axis=0)

        ax.loglog(mean_sfs_unif.index, mean_sfs_unif.values, marker='o', linestyle='-', 
                  label=f'uniformgeo {center}', color=width_colors['uniformgeo'])
        ax.errorbar(mean_sfs_unif.index, mean_sfs_unif.values, yerr=std_sfs_unif.values, fmt='o', 
                    color=width_colors['uniformgeo'], capsize=5, linestyle='None', alpha=0.7)

        ax.set_title(f"SFS for {center}")
        ax.set_xlabel('Allele Count')
        ax.set_ylabel('Count of SNPs')
        ax.grid(False)  # Remove grid lines
        ax.legend(loc='upper right', fontsize=8)  # Add legend in each subplot
        ax.set_xlim(1e0,1e3)
    plt.tight_layout()
    plt.show()

# Process each category and plot SFS for each
plot_sfs_for_scenarios('../results/daf/chr1_lof*center*geo*', fixed_alleles_lof, 'LOF')
plot_sfs_for_scenarios('../results/daf/chr1_mis*center*geo*', fixed_alleles_mis, 'MIS')
plot_sfs_for_scenarios('../results/daf/chr1_syn*center*geo*', fixed_alleles_syn, 'SYN')
```

```python
def plot_sfs_for_scenarios(pattern, fixed_alleles, label, color):
    files = glob.glob(pattern)
    dfs = {file.split('/')[-1]: pd.read_csv(file, delimiter='\t') for file in files}
    
    mean_freqs = {}
    for file, df in dfs.items():
        scenario = file.split('_')[2]
        df_filtered = df[~df['POS'].isin(fixed_alleles)]
        
        # Calculate derived allele counts
        df_filtered['Derived_Alleles_Counts'] = (df_filtered['Derived_Frequency'] * 2 * df_filtered['NOBS']).round().astype(int)
        
        # Count how many SNPs fall into each derived allele count (SFS calculation)
        count_of_alleles = df_filtered['Derived_Alleles_Counts'].value_counts().sort_index()
        
        # Store the SFS counts for this scenario
        mean_freqs[scenario] = count_of_alleles
    
    # Plotting the SFS for each scenario
    plt.figure(figsize=(10, 6))
    
    for scenario, count_of_alleles in mean_freqs.items():
        plt.loglog(count_of_alleles.index, count_of_alleles.values, marker='o', linestyle='-', label=scenario, color=color)
    
    # Final plot adjustments
    plt.title(f"SFS for {label}")
    plt.xlabel('Derived Allele Count')
    plt.ylabel('Count of SNPs')
    plt.legend()
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Process each category and plot SFS for each
plot_sfs_for_scenarios('../results/daf/chr1_lof*geo*', fixed_alleles_lof, 'LOF', 'blue')
plot_sfs_for_scenarios('../results/daf/chr1_mis*geo*', fixed_alleles_mis, 'MIS', 'red')
plot_sfs_for_scenarios('../results/daf/chr1_syn*geo*', fixed_alleles_syn, 'SYN', 'green')
```
