```python
import numpy as  np
import pandas as pd
from matplotlib import pyplot as plt
import glob
from scipy.special import expi
import os.path

plt.rcParams['agg.path.chunksize'] = 200
plt.rcParams.update({'axes.facecolor': 'white', 'figure.facecolor': 'white'})
plt.rcParams.update({'font.size':36})

def get_lc_squared(sigma,s):
    return sigma**2/s

def get_EPsquared(mu,s,rho,sigma,u2):
    lcs = get_lc_squared(sigma,s)
    return (2*mu/(s**2*rho*lcs))*u2+mu**2/s**2
    # return (mu/s)**2 +((mu)/((s**2)*rho*lcs))*u2

def u2_exact(w):
    return (-1*np.exp(w**2/(4*np.pi**2))/(8*np.pi))*expi(-1*w**2/(4*np.pi**2))

mu = 1e-9
sigma = 10
rho = 1
L=10000
N = rho * L**2
```

```python
prefix = "../results/gaussian_w{w}/s{s}_mu{mu}/s{s}_mu{mu}_rho1_L10000_sigma{sigma}_time1000000.0_r0.1_burnin_wrapped_norm_gaussian_w{w}_nside{nside}_{rep}"
w_vals = [100,1000]
nside_vals = [10,100,1000]
sigma_vals = [10,100]
for s in [0.1,0.01]:
    for rep in ["all"]: # [f"rep{i}" for i in range(10)] + 
        for w in w_vals:
            for n in nside_vals:
                for sigma in sigma_vals:
                    pfile = prefix.format(mu=mu, s=s, rho=rho, L=L, rep=rep, w=w, nside=n,sigma=sigma) + ".p"
                    zfile = prefix.format(mu=mu, s=s, rho=rho, L=L, rep=rep, w=w, nside=n,sigma=sigma) + ".zero"

                    try:
                        ps = np.loadtxt(pfile)
                        zeros = np.loadtxt(zfile)
                    except:
                        continue
                    n_samples = np.arange(1, len(ps) + 1)
                    n_zeros = zeros * n_samples / n_samples[-1]
                    correction = (1+np.sqrt(s))/n**2
                    # ps = ps/correction
                    rho = 1
                    w_tilde = w/np.sqrt(get_lc_squared(sigma,s))
                    theory_u2 = u2_exact(w_tilde)
                    theory_EP2 = get_EPsquared(mu,s,rho,sigma,theory_u2) ## FIND BUG HERE
                    
                    print(f"s = {s}")
                    print(f"nside = {n}")
                    print(f"w = {w}")
                    print(f"sigma = {sigma}")
                    print(f"mu / s = {mu / s}")
                    print(f"E[P] = {np.sum(ps) / (len(ps) + n_zeros)}")
                    print(f"E[P^2] = {np.sum(ps**2) / (len(ps) + n_zeros)}")
                    print(f"Theory E[P^2]: {theory_EP2}")
                    print(f"Proportion of sampled p's that are 0: {(len(ps)-np.count_nonzero(ps))/len(ps)}")
                    print(f"Ratio sim to theory E[P]: {(np.sum(ps) / (len(ps) + n_zeros[-1]))/(mu/s)}")
                    print(f"Ratio sim to theory E[P^2]: {(np.sum(ps**2) / (len(ps) + n_zeros[-1]))/(theory_EP2)}")
                    print("-------------------------------")
        
                    plt.figure(figsize=(20,24))
                    plt.subplot(311)
                    plt.plot(N*ps)
                    plt.hlines([1 / s, 2 / s], 0, len(ps), color="black")
                    plt.ylabel("$NP$")
                    # plt.ylim([0, 10/s])
                    plt.title(f"s={s}, w={w}, rep={rep}, nside={n}, sigma={sigma}")
                    plt.subplot(312)
                    plt.plot(np.cumsum(ps) / (n_samples + n_zeros))
                    plt.hlines(mu / s, 0, n_samples[-1], color="black")
                    plt.ylabel("running mean $P$")
                    plt.subplot(313)
                    plt.plot(np.cumsum(ps**2) / (n_samples + n_zeros))
                    
                    
                    plt.hlines(theory_EP2, 0, n_samples[-1], color="black")
                    plt.ylabel("running mean $P^2$")
                    plt.xlabel("sample number")
                    plt.tight_layout()
                    plt.show()
                    # plt.savefig(f'plots_20240415/multi_center_wrapnorm_s{s}_mu{mu}_w{w}_sigma{sigma}_3panel_fixzeros_nside{n}.png')
                    # plt.close()
```
